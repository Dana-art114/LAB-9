```python
cat_feats = ['purpose']
```

```python
final_data = pd.get_dummies(loans, columns=cat_feats, drop_first=True)
```

```python
X = final_data.drop('not.fully.paid', axis=1)
y = final_data['not.fully.paid']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=101
)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train, y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, rfc_pred))
```

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, rfc_pred))
```

```python
from sklearn.metrics import classification_report

print("Decision Tree:\n", classification_report(y_test, predictions))
print("Random Forest:\n", classification_report(y_test, rfc_pred))
```

```python
# Decision Tree performed better for class 1
# because it has higher recall (0.24) compared to Random Forest (0.02)

# Random Forest has higher overall accuracy (0.85 vs 0.73),
# but it fails to correctly predict class 1 (very low recall)

# Therefore, Decision Tree is better for this problem
# since detecting class 1 (not fully paid loans) is more important
```
